## TEXT PREPROCESSING

## CASE FOLDING

```python
import re

# membuat fungsi untuk case folding
def casefolding(text):
    text = text.lower()                                 # mengubah text menjadi huruf kecil
    text = re.sub(r'https?://\S+|www\.S+', '', text)    # menghapus url dari kalimat
    text = re.sub(r'[^\w\s]', '', text)                 # menghapus tanda baca
    text = text.strip()
    return text
```

```python
# membandingkan before dan after casefolding
raw_sample = data['teks'].iloc[2]
case_folding = casefolding(raw_sample)

print('Raw data\t : ',raw_sample)
print('Case Folding\t : ',case_folding)
```

## WORD NORMALIZATION

```python
key_norm = pd.read_csv('key_norm.csv')

def text_normalize(text):
    text = ' '.join([key_norm[key_norm['singkat'] == word]['hasil'].values[0]
    if (key_norm['singkat'] == word).any()
    else word for word in text.split()
    ])

    text = str.lower(text)
    return text
```

```python
# membandingkan before dan after word normalization

raw_data = data['teks'].iloc[2]
word_normal = text_normalize(case_folding)

print('Raw Data\t :',raw_data)
print('Word Normalize\t :',word_normal)
```

## STOPWORD REMOVAL (FILTERING)

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

stopwords_ind = stopwords.words('indonesian')
```

```python
len(stopwords_ind)
```

```python
# melihat daftar stopwords dari nltk
stopwords_ind
```

```python
# membuat fungsi stopword removal

# menambahkan kata ke dalam stopword
more_stopword = ['ya', 'iya', 'nih', 'mas', 'nah', 'piak', 'sih', 'nya', 'serta', 'untuk', 'oh', 'loh', 'mana', 'si', 'eh', 'ln']
stopwords_ind = stopwords_ind + more_stopword

# daftar kata yang tidak boleh dihapus meskipun ada di stopword
exception_words = ['tidak', 'enggak', 'nggak', 'belum']

def remove_stop_word(text):
    clean_words = []
    text = text.split()
    for word in text:
        # hanya hapus jika kata ada di stopword dan bukan exception
        if word not in stopwords_ind or word in exception_words:
            clean_words.append(word)
    return " ".join(clean_words)
```

```python
raw_sample = data['teks'].iloc[6]
case_folding = casefolding(raw_sample)
word_normal = text_normalize(case_folding)
stopword_removal = remove_stop_word(word_normal)

print('Raw Data \t\t :', raw_sample)
print('Case Folding \t\t :', case_folding)
print('Word Normalize \t\t :',word_normal)
print('Stop Word Removal \t\t :', stopword_removal)
```

## TEXT PREPROCESSING PIPELINE

```python
# membuat fungsi untuk menghubungkan seluruh langkah text preprocessing
def text_preprocessing_process(text):
    text = casefolding(text)
    text = text_normalize(text)
    text = remove_stop_word(text)
    text = stemming(text)
    return text
```

```python
%%time
data['clean_teks'] = data['teks'].apply(text_preprocessing_process)
```

```python
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from xgboost import XGBClassifier

# === Data Preparation ===
x = data['clean_teks']
y = data['label']

# === Pipeline: TF-IDF + Chi2 Feature Selection + XGBoost ===
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(ngram_range=(1,2))), # feature extraction
    ('select', SelectKBest(chi2, k=10000)),   # feature selection
    ('clf', XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'))
])

# Split data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Train pipeline
pipeline.fit(x_train, y_train)

# Evaluate
y_pred = pipeline.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

# Save pipeline
with open("pipeline_asesmen.pkl", "wb") as f:
    pickle.dump(pipeline, f)
```
